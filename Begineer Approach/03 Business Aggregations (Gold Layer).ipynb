```python
# Sales by State
from pyspark.sql.functions import sum, count, countDistinct, col, date_format

gold_sales_by_state_df = (
    spark.table("real_time_projects.ecommerce_historical.fact_sales")
    .groupBy("customer_state")
    .agg(
        sum("revenue").alias("total_revenue"),
        countDistinct("order_id").alias("total_orders"),
        count("*").alias("total_items")
    )
    .orderBy(col("total_revenue").desc())
)

display(gold_sales_by_state_df)
```

```python
# Top Products by Revenue
from pyspark.sql.functions import sum, count, countDistinct, col, date_format

gold_top_products_df = (
    spark.table("real_time_projects.ecommerce_historical.fact_sales")
    .groupBy("product_id", "product_category_name")
    .agg(
        sum("revenue").alias("total_revenue"),
        countDistinct("order_id").alias("total_orders"),
        count("*").alias("total_items")
    )
    .orderBy(col("total_revenue").desc())
)

display(gold_top_products_df)
```

```python
# Daily Sales Trend
from pyspark.sql.functions import sum, count, countDistinct, col, date_format

gold_daily_sales_df = (
    spark.table("real_time_projects.ecommerce_historical.fact_sales")
    .groupBy("order_date")
    .agg(
        sum("revenue").alias("daily_revenue"),
        countDistinct("order_id").alias("daily_orders"),
        count("*").alias("daily_items")
    )
    .orderBy("order_date")
)

display(gold_daily_sales_df)
```

```python
# Monthly Sales Trend
from pyspark.sql.functions import sum, count, countDistinct, col, date_format

gold_monthly_sales_df = (
    spark.table("real_time_projects.ecommerce_historical.fact_sales")
    .withColumn("year_month", date_format("order_date", "yyyy-MM"))
    .groupBy("year_month")
    .agg(
        sum("revenue").alias("monthly_revenue"),
        countDistinct("order_id").alias("monthly_orders")
    )
    .orderBy("year_month")
)

display(gold_monthly_sales_df)
```
