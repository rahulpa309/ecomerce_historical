```python
from pyspark.sql.types import *

customer_df = (
    spark.read
    .format("csv")
    .option("header","True")
    .load("/Volumes/real_time_projects/ecommerce_historical/lakehouse_vol/Raw_DS/olist_customers_dataset.csv")
)

order_items_df = (
    spark.read
    .format("csv")
    .option("header","True")
    .load("/Volumes/real_time_projects/ecommerce_historical/lakehouse_vol/Raw_DS/olist_order_items_dataset.csv")
)

order_payments_df = (
    spark.read
    .format("csv")
    .option("header","True")
    .load("/Volumes/real_time_projects/ecommerce_historical/lakehouse_vol/Raw_DS/olist_order_payments_dataset.csv")
)

orders_df = (
    spark.read
    .format("csv")
    .option("header","True")
    .load("/Volumes/real_time_projects/ecommerce_historical/lakehouse_vol/Raw_DS/olist_orders_dataset.csv")
)

products_df = (
    spark.read
    .format("csv")
    .option("header","True")
    .load("/Volumes/real_time_projects/ecommerce_historical/lakehouse_vol/Raw_DS/olist_products_dataset.csv")
)

sellers_df = (
    spark.read
    .format("csv")
    .option("header","True")
    .load("/Volumes/real_time_projects/ecommerce_historical/lakehouse_vol/Raw_DS/olist_sellers_dataset.csv")
)

customer_df.repartition(1) \
    .write \
    .format("parquet") \
    .mode("overwrite") \
    .save("/Volumes/real_time_projects/ecommerce_historical/lakehouse_vol/bronze/customers/")

order_items_df.repartition(1) \
    .write \
    .format("parquet") \
    .mode("overwrite") \
    .save("/Volumes/real_time_projects/ecommerce_historical/lakehouse_vol/bronze/order_items/")

order_payments_df.repartition(1) \
    .write \
    .format("parquet") \
    .mode("overwrite") \
    .save("/Volumes/real_time_projects/ecommerce_historical/lakehouse_vol/bronze/payments/")

orders_df.repartition(1) \
    .write \
    .format("parquet") \
    .mode("overwrite") \
    .save("/Volumes/real_time_projects/ecommerce_historical/lakehouse_vol/bronze/orders/")

products_df.repartition(1) \
    .write \
    .format("parquet") \
    .mode("overwrite") \
    .save("/Volumes/real_time_projects/ecommerce_historical/lakehouse_vol/bronze/products/")

sellers_df.repartition(1) \
    .write \
    .format("parquet") \
    .mode("overwrite") \
    .save("/Volumes/real_time_projects/ecommerce_historical/lakehouse_vol/bronze/sellers/")
```
